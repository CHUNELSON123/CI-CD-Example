```python
from google.colab import files
uploaded = files.upload()

!pip install pandas numpy matplotlib seaborn scikit-learn tensorflow

# 1. IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# 2. LOAD DATA
df = pd.read_csv('Bambili_IrrPT.csv')

# Fix column names
df.columns = [col.lower().strip().replace(' ', '_') for col in df.columns]  # clean names
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Show first rows
df.head()

# 3. FEATURE SELECTION
features = ['temperature', 'humidity', 'wind_speed', 'potential']
target = 'irradiance'

# 4. SCALING
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[features + [target]])
scaled_df = pd.DataFrame(scaled_data, columns=features + [target], index=df.index)

# 5. SEQUENCE CREATION FUNCTION
def create_sequences(data, seq_len=7):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len, :-1])  # inputs
        y.append(data[i+seq_len, -1])     # target
    return np.array(X), np.array(y)

SEQ_LEN = 7
X, y = create_sequences(scaled_data, SEQ_LEN)

# 6. SPLIT DATA
train_size = int(0.7 * len(X))
val_size = int(0.15 * len(X))

X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:train_size+val_size], y[train_size:train_size+val_size]
X_test, y_test = X[train_size+val_size:], y[train_size+val_size:]

# 7. LSTM MODEL
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(SEQ_LEN, X.shape[2])),
    Dropout(0.2),
    LSTM(64),
    Dropout(0.2),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# 8. TRAIN MODEL
history = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                    epochs=50, batch_size=32,
                    callbacks=[EarlyStopping(patience=5, restore_best_weights=True)])

# 9. PREDICT
y_pred = model.predict(X_test)

# 10. INVERSE SCALE
# Pad other features with 0s to inverse transform correctly
irr_index = features + [target].index(target)
y_test_actual = scaler.inverse_transform(
    np.hstack([np.zeros((len(y_test), len(features))), y_test.reshape(-1,1)])
)[:, -1]

y_pred_actual = scaler.inverse_transform(
    np.hstack([np.zeros((len(y_pred), len(features))), y_pred])
)[:, -1]

# 11. EVALUATE
mae = mean_absolute_error(y_test_actual, y_pred_actual)
rmse = mean_squared_error(y_test_actual, y_pred_actual, squared=False)
print(f"\nüìä MAE: {mae:.2f}")
print(f"üìâ RMSE: {rmse:.2f}")

# 12. PLOT
plt.figure(figsize=(12,5))
plt.plot(y_test_actual, label='Actual Irradiance')
plt.plot(y_pred_actual, label='Predicted Irradiance')
plt.title('LSTM Forecasting: Bambili Solar Irradiance')
plt.xlabel('Days')
plt.ylabel('Irradiance (W/m¬≤)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
 # 1. IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# 2. UPLOAD DATA
from google.colab import files
uploaded = files.upload()  # Upload Bambili_IrrPT.csv

# 3. LOAD & CLEAN DATA
df = pd.read_csv('Bambili_IrrPT.csv')
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]  # clean headers
print("‚úÖ Cleaned columns:", df.columns.tolist())

df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# 4. SELECT FEATURES & TARGET
features = ['temperature', 'humudity', 'wind_speed', 'potential']
target = 'irradiance'

# 5. SCALE DATA
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[features + [target]])
scaled_df = pd.DataFrame(scaled_data, columns=features + [target], index=df.index)

# 6. CREATE SEQUENCES
def create_sequences(data, seq_len=7):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len, :-1])
        y.append(data[i+seq_len, -1])
    return np.array(X), np.array(y)

SEQ_LEN = 7
X, y = create_sequences(scaled_data, SEQ_LEN)

# 7. SPLIT INTO TRAIN / VAL / TEST
train_size = int(0.7 * len(X))
val_size = int(0.15 * len(X))

X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:train_size+val_size], y[train_size:train_size+val_size]
X_test, y_test = X[train_size+val_size:], y[train_size+val_size:]

# 8. DEFINE LSTM MODEL
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(SEQ_LEN, X.shape[2])),
    Dropout(0.2),
    LSTM(64),
    Dropout(0.2),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# 9. TRAIN MODEL
history = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                    epochs=50, batch_size=32,
                    callbacks=[EarlyStopping(patience=5, restore_best_weights=True)])

# 10. MAKE PREDICTIONS
y_pred = model.predict(X_test)

# 11. INVERSE SCALING FOR EVALUATION
y_test_actual = scaler.inverse_transform(
    np.hstack([np.zeros((len(y_test), len(features))), y_test.reshape(-1,1)])
)[:, -1]

y_pred_actual = scaler.inverse_transform(
    np.hstack([np.zeros((len(y_pred), len(features))), y_pred])
)[:, -1]

# 12. EVALUATION METRICS
mae = mean_absolute_error(y_test_actual, y_pred_actual)
rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred_actual))

print(f"\nüìä MAE: {mae:.2f}")
print(f"üìâ RMSE: {rmse:.2f}")


# 13. PLOT RESULTS
plt.figure(figsize=(12,5))
plt.plot(y_test_actual, label='Actual Irradiance')
plt.plot(y_pred_actual, label='Predicted Irradiance')
plt.title('LSTM Forecasting: Bambili Solar Irradiance')
plt.xlabel('Days')
plt.ylabel('Irradiance (W/m¬≤)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
